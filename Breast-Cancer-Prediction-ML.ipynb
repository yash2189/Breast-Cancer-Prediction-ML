```python
from sklearn.preprocessing import LabelEncoder
```

```python
labelencoder_Y = LabelEncoder()
Y = labelencoder_Y.fit_transform(y)
```

```python
Y #M-1 , B-0
```

```python
# Splitting the dataset into the Training set and Test set
from sklearn.model_selection import train_test_split
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size = 0.35, random_state = 0)
```

```python
#Feature Scaling
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
X_train = sc.fit_transform(X_train)
X_test = sc.transform(X_test)
```

```python
#Importing Models
from sklearn.linear_model import LogisticRegression
from sklearn.neighbors import KNeighborsClassifier
from sklearn.svm import SVC
from sklearn.naive_bayes import GaussianNB
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn import metrics
```

```python
import warnings
warnings.filterwarnings('ignore')
```

```python
def predict(classifier):
    classifier.fit(X_train,Y_train)
    y_pred=classifier.predict(X_test)
    return y_pred
```

```python
def model_score(Result):
    score = metrics.accuracy_score(Y_test, Result)
    #confusion_matrix = metrics.confusion_matrix(Y_test,Result)
    return score
```

```python
classifier_1 = LogisticRegression(random_state = 0)
Result_1=predict(classifier_1)
Accuracy_1 = (model_score(Result_1))*100
```

```python
classifier_2 = KNeighborsClassifier(n_neighbors = 5, metric = 'minkowski', p = 2)
Result_2=predict(classifier_2)
Accuracy_2 = (model_score(Result_2))*100
```

```python
classifier_3 = SVC(kernel = 'linear', random_state = 0)
Result_3=predict(classifier_3)
Accuracy_3 = (model_score(Result_3))*100
```

```python
classifier_4 = SVC(kernel = 'rbf', random_state = 0)
Result_4=predict(classifier_4)
Accuracy_4 = (model_score(Result_4))*100
```

```python
classifier_5=GaussianNB()
Result_5=predict(classifier_5)
Accuracy_5 = (model_score(Result_5))*100
```

```python
classifier_6=DecisionTreeClassifier(criterion = 'entropy', random_state = 0)
Result_6=predict(classifier_6)
Accuracy_6 = (model_score(Result_6))*100
```

```python
classifier_7 = RandomForestClassifier(n_estimators = 10, criterion = 'entropy', random_state = 0)
Result_7=predict(classifier_7)
Accuracy_7 = (model_score(Result_7))*100
```

```python
models = pd.DataFrame({'Model Name':["Logistic Regression","KNN","SVC Linear","SVC RBF","Naive Bayes","Decision Tree","Random Forest"],\
                        'Accuracy':[Accuracy_1,Accuracy_2,Accuracy_3,Accuracy_4,Accuracy_5,Accuracy_6,Accuracy_7]})
```
